```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("train.csv")

# Display basic info
print("Basic Info:\n")
print(df.info())
print("\nSummary Statistics:\n")
print(df.describe())
print("\nColumn Names:\n", df.columns)

# Value counts for categorical columns
print("\nSurvival Count:\n", df['Survived'].value_counts())
print("\nClass Count:\n", df['Pclass'].value_counts())
print("\nSex Count:\n", df['Sex'].value_counts())
print("\nEmbarked Count:\n", df['Embarked'].value_counts())

# Plot 1: Histogram of Age
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Plot 2: Boxplot of Fare
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.tight_layout()
plt.show()

# Plot 3: Countplot of Survival by Sex
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival Count by Sex")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Plot 4: Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

# Plot 5: Pairplot (cleaned)
df_pairplot = df[['Survived', 'Pclass', 'Age', 'Fare']].dropna()
sns.pairplot(df_pairplot, hue='Survived', diag_kind='hist')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='Set1')
plt.title('Age vs Fare by Survival')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Pclass', hue='Survived', palette='coolwarm')
plt.title('Age vs Passenger Class by Survival')
plt.xlabel('Age')
plt.ylabel('Pclass')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Fare', y='Pclass', hue='Sex')
plt.title('Fare vs Passenger Class by Sex')
plt.xlabel('Fare')
plt.ylabel('Pclass')
plt.tight_layout()
plt.show()
```
